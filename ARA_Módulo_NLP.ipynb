## Definición NLP

El **procesamiento del lenguaje natural (NLP)** es un campo interdisciplinar que une la lingüística, la informática y la inteligencia artificial con el objetivo de programar ordenadores para **procesar y analizar grandes cantidades de datos del lenguaje natural**.

El objetivo final es que los **ordenadores puedan "entender" el contenido de los documentos**, incluyendo los matices contextuales del lenguaje que contienen, para extraer información y clasificar y organizar los documentos.

Los principales retos que plantea el procesamiento del lenguaje natural son el reconocimiento del habla, la comprensión del lenguaje natural y la generación de lenguaje natural.

## Etapas de un proyecto de NLP

Las etapas de un proyecto de Procesamiento del Lenguaje Natural (NLP) pueden variar según el alcance y los objetivos del proyecto.

Muchas de las etapas son las mismas que para cualquier otro proyecto de machine learning.

Los únicos cambios que hay es a la hora de procesar y limpiar los datos.

Sin embargo, en términos generales, las siguientes etapas son comunes en la mayoría de los proyectos de NLP:

1. **Definición del problema y objetivos :**
   - Tener claro el problema que se quiere resolver
   - Establecer claramente los objetivos del proyecto y determinar qué tipo de tareas de NLP se realizarán.
   - Definir el alcance del proyecto para evitar desviaciones innecesarias.

2. **Obtención de Datos:**
   - Obtener el conjunto de datos relevante para el proyecto.
   - Estos datos pueden ser texto escrito, transcripciones de voz, etiquetas de lenguaje, etc.

3. **Preprocesamiento de Texto:**
   - Limpiar y preparar los datos antes del análisis.
   - Esto incluye tareas como eliminación de caracteres especiales, tokenización, eliminación de stop words, lematización, etc.

4. **Representación del texto en vectores**
   - Es necesario convertir los textos a algo que un ordenador entienda, es decir, números.
   - Normalmente se suelen representar como vectores
   - Aquí entran en juego el Word2Vec, Word Embeddings o TF-IDF.

5. **Exploración y Análisis de Datos:**
   - Realizar un análisis exploratorio de los datos para comprender mejor las características del texto, identificar patrones y tendencias relevantes.

6. **Aplicación de NLP:**
   - Determinar qué técnicas y modelos de NLP son apropiados para abordar los objetivos del proyecto.
   - Esto puede incluir análisis de sentimientos, clasificación de texto, modelado de tópicos, etc.

7. **Entrenamiento y Evaluación de Modelos:**
   - Entrenar los modelos de NLP utilizando técnicas de aprendizaje automático o procesamiento estadístico.
   - Evaluar el rendimiento de los modelos con conjuntos de datos de prueba.

8. **Ajuste y Optimización:**
   - Realizar ajustes y optimizaciones en los modelos y parámetros para mejorar el rendimiento y la precisión del proyecto.

9. **Implementación y Despliegue:**
   - Implementar el modelo o sistema de NLP en producción, lo que puede implicar la integración con sistemas existentes o la creación de una interfaz para el usuario.

10. **Pruebas y Validación:**
   - Realizar pruebas exhaustivas para asegurarse de que el proyecto funcione correctamente y cumpla con los requisitos y objetivos establecidos.

11. **Documentación y Presentación:**
    - Documentar todo el proceso y los resultados del proyecto. Presentar los hallazgos y resultados a los interesados o a un público relevante.

12. **Mantenimiento y Mejora Continua:**
    - Realizar el mantenimiento del proyecto y continuar mejorando el modelo o sistema a medida que se recopilan más datos o se encuentran nuevas oportunidades de mejora.

Estas etapas proporcionan una guía general para llevar a cabo un proyecto de NLP con éxito.

Es importante adaptar y ajustar estas etapas según las necesidades específicas de cada proyecto y los recursos disponibles.

## Limpieza de textos - Regex

Las **expresiones regulares (regex)** son patrones de texto para buscar o reemplazar caracteres o secuencias de caracteres en un texto.

La **limpieza de texto con regex** se utiliza para **eliminar información no deseada o ruidosa** que puede afectar la calidad y precisión del análisis.

Algunos ejemplos de limpieza de texto con regex incluyen:

1. **Eliminación de caracteres especiales:** Buscar y eliminar caracteres especiales como signos de puntuación, símbolos, números, etc., que no aportan valor al análisis y pueden dificultar la tokenización y el procesamiento posterior.

2. **Normalización de texto:** Convertir el texto a minúsculas para evitar problemas de casamiento en el análisis. También se puede realizar una normalización adicional, como eliminar acentos o caracteres diacríticos para estandarizar el texto.

3. **Eliminación de URLs:** En caso de que el texto contenga enlaces web, se pueden utilizar expresiones regulares para detectar y eliminar las URLs, ya que generalmente no son relevantes para el análisis.

4. **Eliminación de etiquetas HTML:** Si el texto proviene de contenido web, es posible que contenga etiquetas HTML que deben eliminarse antes del análisis.

5. **Extracción de menciones y hashtags:** En el análisis de redes sociales, puede ser útil extraer menciones de usuarios y hashtags para su posterior análisis.

6. **Reemplazo de contracciones:** Se pueden utilizar expresiones regulares para reemplazar contracciones comunes, como "don't" por "do not", para estandarizar el texto.

**Ejemplo limpieza texto con Python**

- Se utilizará el módulo re
- Se trabajará en un texto con caracteres especiales, números y símbolos
- El objetivo será dejar el texto limpio
- Se utilizarán regex para eliminar esos caracteres

```python
# importamos módulo re
import re

# creamos texto con caracteres especiales
texto = "¡Hola! Este es un ejemplo de limpieza de texto con expresiones regulares en Python. 😃 #NLP #regex 12345"
```

Ahora limpiamos el texto con regex

```python
# Eliminar caracteres especiales y símbolos
texto_limpio = re.sub(r'[^\w\s]', '', texto)

# Eliminar números
texto_limpio = re.sub(r'\d+', '', texto_limpio)

# Eliminar hashtags
texto_limpio = re.sub(r'#\w+', '', texto_limpio)

print(texto_limpio)
```

En este ejemplo, hemos utilizado las siguientes expresiones regulares:

- **r'[^\w\s]'**: Corresponde a cualquier caracter que no sea una letra, un número o un espacio en blanco. Al usar re.sub() con esta expresión regular, hemos eliminado los caracteres especiales y los emojis (😃).

- **r'\d+'**: Corresponde a uno o más dígitos numéricos. Al usar re.sub() con esta expresión regular, hemos eliminado los números (12345).

- **r'#\w+'**: Corresponde a cualquier palabra que comience con el símbolo #, es decir, los hashtags. Al usar re.sub() con esta expresión regular, hemos eliminado los hashtags (#NLP y #regex).

**Nota:** Las expresiones regulares pueden variar según los requisitos específicos de limpieza de texto de cada proyecto, por lo que es posible que necesites ajustarlas para satisfacer tus necesidades particulares.

## Tokenization

La tokenización consiste en **dividir un texto en unidades más pequeñas llamadas "tokens"**.

Un **token** puede ser una **palabra, una frase o incluso un carácter**, dependiendo del nivel de detalle deseado en el análisis.

Es el **primer paso en el procesamiento de texto** y es esencial para el análisis y comprensión del lenguaje humano.

Puede ser **útil en varias aplicaciones de NLP**, como análisis de sentimientos, clasificación de texto, extracción de información, generación de lenguaje natural y más.

Al **dividir el texto en tokens**, se puede analizar cada unidad de manera individual, lo que **facilita la manipulación y el procesamiento del lenguaje**.

```python
# Ejemplo con Python y SpaCy:

import spacy

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

# Definir una función para realizar la tokenización
def tokenizacion(texto):
    doc = nlp(texto)
    tokens = [token.text for token in doc]
    return tokens

texto_tokenizado = tokenizacion(texto_limpio)

print(texto_tokenizado)
```

En este ejemplo, hemos utilizado el modelo de lenguaje en español de spaCy para realizar la tokenización en el texto dado. La función `tokenizacion()` toma el texto como entrada y devuelve una lista de tokens. Cada elemento de la lista representa una palabra o símbolo individual del texto original.

La tokenización es un paso fundamental en el procesamiento de texto en NLP, ya que es el punto de partida para muchas otras tareas de análisis y modelado de lenguaje. Al dividir el texto en unidades más pequeñas, se obtiene una representación estructurada que facilita el análisis y la manipulación del lenguaje humano.

## Stopwords removal

Las "**stopwords**" (palabras vacías o de parada) son **palabras** que se consideran **comunes y poco informativas** para el análisis de texto debido a su alta frecuencia en el lenguaje natural.

Estas palabras suelen ser **pronombres, preposiciones, conjunciones y otros términos muy frecuentes** que no aportan mucho significado al contexto del texto.

La **eliminación de stopwords** permite **reducir el tamaño del vocabulario y mejorar la eficiencia y precisión del análisis.**

Al eliminar las stopwords, se **reduce el ruido** y se d**estacan las palabras clave** y significativas que realmente aportan información relevante para las tareas de NLP.

image.png

```python
# Ejemplo stopwords removal en Python
import spacy

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

# Definir una función para eliminar stopwords
def eliminar_stopwords(texto):
    doc = nlp(texto)
    tokens_sin_stopwords = [token.text for token in doc if not token.is_stop]
    texto_limpio = " ".join(tokens_sin_stopwords)
    return texto_limpio

texto_resultado = eliminar_stopwords(texto_limpio)

print(texto_resultado)
```

## Stemming y Lemmatization

El **"stemming"** consiste en reducir las palabras a su **forma raíz** o "stem".

El objetivo del stemming es **eliminar los sufijos y prefijos de las palabras** para agruparlas en su forma base y reducir la variación morfológica.

El stemming es útil para **simplificar el análisis de texto** y **reducir la dimensionalidad del vocabulario.**

Al reducir las palabras a su forma raíz, se agrupan palabras relacionadas que comparten una raíz común.

Sin embargo, es importante tener en cuenta que el stemming **puede producir raíces que no son palabras válidas o significativas en sí mismas**, ya que no tiene en cuenta el contexto.

image.png

La **lematización** consiste en **reducir las palabras a su forma base o "lemas"**.

A diferencia del stemming, que simplemente corta los sufijos de las palabras para obtener una raíz, **la lematización tiene en cuenta el significado y contexto** de las palabras para obtener su forma base, que es una palabra válida y significativa en el idioma.

El **lema** de una palabra es **su forma canónica** o diccionario.

Por ejemplo, el lema del verbo "corriendo" es "correr", y el lema del sustantivo "gatos" es "gato".

La lematización es especialmente útil en tareas de análisis de texto donde se requiere una mayor precisión y se quiere reducir la variación morfológica de las palabras.

## POS

**POS (Part-of-Speech)** se refiere a la tarea de **etiquetar cada palabra de un texto con su categoría gramatical correspondiente**, como sustantivos, verbos, adjetivos, pronombres, preposiciones, etc.

Estas categorías gramaticales proporcionan **información** importante sobre **cómo se relacionan las palabras en una oración** y son fundamentales para el análisis y comprensión del lenguaje humano.

Ayuda a **entender la estructura y el significado de las oraciones**, y también es útil para realizar tareas más complejas, como el **análisis sintáctico y la extracción de información**.

```python
import spacy

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

# Definir una función para realizar el etiquetado POS
def etiquetado_pos(texto):
    doc = nlp(texto)
    pos_tags = [(token.text, token.pos_) for token in doc]
    return pos_tags

texto = "La casa es grande y bonita."
etiquetas_pos = etiquetado_pos(texto)

print(etiquetas_pos)
```

## NER

**NER (Named Entity Recognition)** o reconocimiento de entidades nombradas es una tarea que consiste en **identificar y clasificar entidades con nombres propios en un texto**, como personas, lugares, organizaciones, fechas, cantidades, etc.

El objetivo del NER es **extraer información relevante y significativa del texto** al identificar y etiquetar estas entidades.

Al identificar las entidades en el texto, **se puede comprender mejor el contenido y estructurar la información** para su posterior procesamiento.

```python
import spacy

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

# Definir una función para realizar el reconocimiento de entidades nombradas (NER)
def reconocimiento_entidades(texto):
    doc = nlp(texto)
    entidades = [(ent.text, ent.label_) for ent in doc.ents]
    return entidades

texto = "El presidente Joe Biden se reunió con la canciller Angela Merkel en Berlín el 25 de agosto de 2023."

entidades_nombradas = reconocimiento_entidades(texto)

print(entidades_nombradas)
```

## Bag of words

**Bag of Words (BoW) o "bolsa de palabras"** es una técnica de representación de texto que **convierte un documento o un conjunto de documentos de texto en un vector de características**.

En BoW, se **crea un "saco" que contiene todas las palabras únicas del texto** (sin repetición) y se cuenta **la frecuencia** de cada palabra en el documento.

El orden y la estructura de las palabras se ignoran en esta técnica, y solo se tiene en cuenta la presencia y frecuencia de cada palabra en el texto.

Sin embargo, esta técnica tiene algunas limitaciones, como la pérdida de información sobre la estructura gramatical y el contexto de las palabras.

image.png

```python
import spacy
from collections import Counter

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

def bag_of_words(texto):
    # Procesar el texto con spaCy
    doc = nlp(texto)

    # Crear una bolsa de palabras utilizando un diccionario (word_freq)
    word_freq = Counter()
    for token in doc:
        if not token.is_stop and not token.is_punct:
            word_freq[token.lemma_] += 1

    return word_freq

# Ejemplo de uso
texto = "Este es un ejemplo de representación Bag of Words con SpaCy. Bag of Words es una técnica de NLP."
bow = bag_of_words(texto)
print(bow)
```

## Bag of n-grams

**El Bag of N-grams (BoN) o "bolsa de N-gramas"** es una extensión del concepto de Bag of Words (BoW) en el Procesamiento del Lenguaje Natural (NLP).

En lugar de considerar palabras individuales como en BoW, BoN **toma en cuenta secuencias contiguas de N palabras llamadas "N-gramas"**.

Un **N-grama** es una **secuencia de N elementos**, que en el caso de BoN son palabras.

El objetivo de BoN es **capturar más información contextual al considerar la co-ocurrencia de palabras en secuencia.**

Por ejemplo:
- N=2 (bigramas)
- Texto: `"Me gusta el helado"`
- Salida bigramas: `"Me gusta", "gusta el" y "el helado"`.



image.png

```python
import spacy
from collections import Counter

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

def bag_of_ngrams(texto, n):
    # Procesar el texto con spaCy
    doc = nlp(texto)

    # Crear una bolsa de n-gramas utilizando un diccionario (ngram_freq)
    ngram_freq = Counter()
    for i in range(len(doc) - n + 1):
        ngram = tuple([token.lemma_ for token in doc[i:i+n]])
        ngram_freq[ngram] += 1

    return ngram_freq

# Ejemplo de uso
texto = "Este es un ejemplo de representación Bag of N-grams con SpaCy. Bag of N-grams es una técnica de NLP."
ngramas = bag_of_ngrams(texto, 2)
print(ngramas)
```

## Word vector o Word Embeddings

La vectorización de textos, también conocida como **Word Vectors** o **Word Embeddings**, es una técnica en el Procesamiento del Lenguaje Natural (NLP) que **asigna palabras a vectores numéricos** en un espacio vectorial continuo.

**Cada palabra se representa como un vector** en ese espacio, lo que **permite capturar las relaciones semánticas y contextuales entre palabras.**

Esta representación numérica es esencial para **permitir que los algoritmos de aprendizaje automático procesen y comprendan el lenguaje humano.**

Estos vectores de palabras, con sus distancias y direcciones, pueden reflejar **relaciones semánticas como similitud o analogía entre palabras**.



image.png

```python
import spacy

# Cargar el modelo de lenguaje en español de spaCy
nlp = spacy.load("es_core_news_sm")

# Procesar un texto con spaCy para obtener vectores de palabras
texto = "El perro corre en el parque."
doc = nlp(texto)

# Imprimir los vectores de palabras para cada token en el texto
for token in doc:
    print(f"Palabra: {token.text}, Vector: {token.vector}")
```

## CountVectorizer

**CountVectorizer** es una clase en la biblioteca Scikit-Learn (Sklearn) que se utiliza para **convertir una colección de documentos de texto en una matriz de recuento de términos**.

Esta matriz representa la **frecuencia de cada término (palabra) en los documentos** y se utiliza comúnmente en tareas de análisis de texto y minería de texto.

La idea detrás de CountVectorizer es que cada **fila de la matriz representa un documento** y cada **columna representa un término único**.

Los valores en la matriz son las frecuencias de los términos en los documentos correspondientes.

Esto permite a los algoritmos de aprendizaje automático trabajar con datos textuales de una manera numérica.

Ejemplo de uso de CountVectorizer con Scikit-Learn:

image.png

```python
# Ejemplo CountVectorizer sklearn
from sklearn.feature_extraction.text import CountVectorizer

# Ejemplo de documentos
documentos = [
    "Este es un documento de ejemplo.",
    "Otro documento para ilustrar el proceso CountVectorizer.",
    "CountVectorizer es una herramienta en Sklearn.",
    "El análisis de texto utiliza CountVectorizer para realizar el análisis."
]

# Crear un objeto CountVectorizer
count_vectorizer = CountVectorizer()

# Ajustar y transformar los documentos con CountVectorizer
matriz_conteo = count_vectorizer.fit_transform(documentos)

# Obtener nombres de las características (palabras)
features = count_vectorizer.get_feature_names_out()

# Convertir la matriz sparse a un array
matriz_conteo_array = matriz_conteo.toarray()

# Imprimir los nombres de las características y la matriz de recuento
print("Características (Palabras):", features)
print("Matriz de Recuento:")
print(matriz_conteo_array)
```

## TF-IDF

**TF-IDF (Term Frequency-Inverse Document Frequency)** es una técnica utilizada en el Procesamiento del Lenguaje Natural (NLP) para **evaluar la importancia relativa de una palabra en un documento dentro de una colección de documentos.**

La idea detrás de TF-IDF es **destacar las palabras que son frecuentes en un documento** pero **raras en el conjunto completo de documentos**, lo que puede ayudar a **identificar términos clave y relevantes** para un documento específico.

El `cálculo de TF-IDF` involucra dos componentes:

- **Term Frequency (TF)**: Mide la **frecuencia de una palabra en un documento específico**. Se calcula como el número de veces que una palabra aparece en el documento dividido por el total de palabras en el documento.

- **Inverse Document Frequency (IDF)**: **Mide la rareza de una palabra en el conjunto de documentos**. Se calcula como el logaritmo del total de documentos dividido por el número de documentos que contienen la palabra.

El producto de TF y IDF produce el valor de TF-IDF para cada palabra en un documento.



image.png

**Nota:** Spacy no tiene un TF-IDF como tal. Usaremos Scikit-Learn

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Ejemplo de documentos
documentos = [
    "Este es un documento de ejemplo.",
    "Otro documento para ilustrar el proceso TF-IDF.",
    "TF-IDF es una técnica importante en NLP.",
    "El análisis de texto se beneficia de TF-IDF."
]

# Crear un objeto TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Ajustar y transformar los documentos con TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(documentos)

# Obtener nombres de las características (palabras)
features = tfidf_vectorizer.get_feature_names_out()

# Convertir la matriz sparse a un array
tfidf_array = tfidf_matrix.toarray()

# Imprimir los nombres de las características y las matrices TF-IDF
print("Características (Palabras):", features)
print("Matriz TF-IDF:")
print(tfidf_array)
```
